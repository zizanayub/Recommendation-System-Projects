### Popularity Based Recommender System

```python
books_with_ratings = ratings.merge(books, on = "ISBN")
```

```python
books_with_ratings.groupby('Book-Title').count()
```

```python
#Grouping books based on count of ratings

num_ratings = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_ratings
```

```python
# Grouping the books based on avg. rating

avg_rating = books_with_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={'Book-Rating':'Avg. Rating'},inplace=True)
avg_rating
```

```python
#merging both num_ratings and Avg. Rating
popular_df = num_ratings.merge(avg_rating)
popular_df
```

```python
#selecting the books with 250 or more than 250 ratings
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values(['Avg. Rating'],ascending=False).head(50)
```

```python
popular_df
```

```python
#merging data based on book information

popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','ISBN','Book-Author','Image-URL-M','Avg. Rating','num_ratings']]
```

```python
popular_df
```

### Collaborative Filtering Based Recommender System

```python
x = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 200
exact_readers = x[x].index
```

```python
filtered_rating = books_with_ratings[books_with_ratings['User-ID'].isin(exact_readers)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
#importing cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
        index = np.where(pt.index==book_name)[0][0]
        similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True) [1:6]

        for i in similar_items:
            print(pt.index[i[0]])
```

```python
recommend('The Notebook')
```
